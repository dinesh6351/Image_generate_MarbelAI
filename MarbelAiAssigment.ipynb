```python
import os
from rembg import remove
from PIL import Image
```

```python
# Function to remove white background from product images
def remove_background(input_path, output_path):
    with open(input_path, "rb") as f:
        img = remove(f.read())
    with open(output_path, "wb") as f:
        f.write(img)

# Define paths for input product images and output advertisement images
input_product_folder = "input_product"
output_advertisement_folder = "output_advertisement"

# Create output advertisement folder if it doesn't exist
if not os.path.exists(output_advertisement_folder):
    os.makedirs(output_advertisement_folder)

# Generate background images 
backgrounds_folder = "background_images"
background_images = [f for f in os.listdir(backgrounds_folder) if os.path.isfile(os.path.join(backgrounds_folder, f))]

# Process each product image
for product_image_name in os.listdir(input_product_folder):
# product img name change
    product1=product_image_name.replace('.jpg','')
    outputjoin=os.path.join(output_advertisement_folder, product1)
    if not os.path.exists(outputjoin):
     os.makedirs(outputjoin)

    # Exclude directories
    if os.path.isdir(os.path.join(input_product_folder, product_image_name)):
        continue

    product_image_path = os.path.join(input_product_folder, product_image_name)
    # Remove white background from product image
    product_image_no_bg_path = os.path.join(outputjoin, "no_bg_" + product_image_name)
    remove_background(product_image_path, product_image_no_bg_path)

    # Merge product image with different background images
    for i, background_image_name in enumerate(background_images):
        background_image_path = os.path.join(backgrounds_folder, background_image_name)
        background = Image.open(background_image_path).convert("RGBA")
        product_no_bg = Image.open(product_image_no_bg_path).convert("RGBA")
        # Resize background image to match product image size
        background = background.resize(product_no_bg.size, Image.BICUBIC)
        # Composite product image with background
        advertisement_image = Image.alpha_composite(background, product_no_bg)
        # Save output advertisement image
        output_image_path = os.path.join(outputjoin, f"{product_image_name[:-4]}_bg_{i+1}.png")
        advertisement_image.save(output_image_path)

print("Advertisement images generation completed!")
```
